### dealing with NA values

```python
nan_features = [features for features in master.columns if master[features].isnull().sum()>=1]
master[nan_features].isnull().sum()
```

```python
# Check NAN Values are important for Target Feature.
trainnan_features = [features for features in train.columns if train[features].isnull().sum()>=1]

train_copy = train.copy()

for feature in trainnan_features:
    train_copy[feature] = np.where(train_copy[feature].isnull(),1,0)
    train_copy.groupby(feature)["SalePrice"].median().plot.bar(color=["pink","b"],figsize=(5,5))
    plt.title(feature)
    plt.show()
```


## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score ,mean_absolute_error,mean_squared_error
from math import sqrt
```

```python
X_train, X_test, y_train, y_test= train_test_split(df_x,df_y,test_size=0.20)
```

```python
X_train
```

```python
model1=DecisionTreeRegressor()
parameters={"max_depth":[10,30,40,50,55,60,70]}
grid1=GridSearchCV(model1, param_grid=parameters, n_jobs=-1,cv=10,verbose=2)
grid1.fit(X_train,y_train)
```

```python
grid1.best_params_
```

```python
y_pred=grid1.predict(X_test)
```

```python
from sklearn.metrics import r2_score ,mean_absolute_error,mean_squared_error
from math import sqrt
```

```python
mse=mean_squared_error(y_test,y_pred)
print("RMSE:",sqrt(mse))
```

```python
model2=RandomForestRegressor()
parameters={"n_estimators":[100,500,700,100], "max_depth":[10,30,40,50,55,60,70]}
grid2=GridSearchCV(model2, param_grid=parameters, n_jobs=-1,cv=10,verbose=2)
grid2.fit(X_train,y_train)
```

```python
y_pred=grid2.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print("RMSE:",sqrt(mse))
```

```python
xgb_model = XGBRegressor()
xgb_model.fit(X_train,y_train)
xgb_model.score(X_train,y_train)
```

```python
xgb_prediction = xgb_model.predict(X_test)
r2_score(y_test,xgb_prediction)
```
