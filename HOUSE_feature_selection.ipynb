```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step Forward Feature Selection

```python
url_data="https://raw.githubusercontent.com/imsy875/House-Prices-Advanced-/main/house_data(Scaled).csv"
url_target="https://raw.githubusercontent.com/imsy875/House-Prices-Advanced-/main/target.csv"
```

```python
data=pd.read_csv(url_data)
y=pd.read_csv(url_target)
```

```python
data
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
model2=RandomForestRegressor(max_depth=45)
```

```python
feature_selection_1 = SFS(model2,k_features=20,verbose=2,cv=5,n_jobs=-1)
feature_selection_1.fit(data,y)
```

```python
feature_selection_1.k_feature_names_
```

### Constant Remove Method

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
constant_filter = VarianceThreshold(threshold=0)
x_df_filter = constant_filter.fit_transform(data)
x_df_filter
```

```python
constant_filter.get_support().sum()
```

```python
constant_columns = [column for column in data.columns if column not in data.columns[constant_filter.get_support()]]
constant_columns
```

###  Backword selection 

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
```

```python
parameters = {"n_estimators":[1000],"max_depth":[40,55],"learning_rate":[0.05],
              "reg_alpha":[0.08, 0.12],"reg_lambda":[30]}
xgb_model=XGBRegressor
grid_xgb = GridSearchCV(xgb_model,param_grid=parameters,n_jobs=-1,cv=10,verbose=2)
```

```python
feature_selection_all = SFS(grid_xgb,k_features=(30),verbose=2,cv=4,n_jobs=-1)
feature_selection_all.fit(data,y)
```

```python
feature_selection_all.k_feature_names_
```
